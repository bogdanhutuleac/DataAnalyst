```python

def line_convert(string):
  for sign in ["(",")"]:
    string = string.replace(sign,"")
  string = string.split("\n")
  string[0] = string[0].replace(" ", "-").lower()
  if string[1] == "Accelerator / Incubator":
    string[1] = "accelerator"
  else:
    string[1] = string[1].lower()

  return tuple(string)
line_convert('''Questo
(Startup)''')
```

```python
# define the lists:
title, status, city, country, content, founders, webaddress, employees, looking_for, categorys, descriptions = ([] for i in range(11))

# scrap https://www.romanianstartups.com for all data about companies
for page in items:
  base_url = f"https://www.romanianstartups.com/{page[1]}/{page[0]}/"
  res = requests.get(base_url)
  soup = bs4.BeautifulSoup(res.text,"lxml")
  try:
    title.append(soup.find("div", class_="t").text)
  except:
    title.append("null")
    
  try:
    status.append(soup.find("div", class_="status").text)
  except:
    status.append("null")
    
  try:
    city.append(soup.find("span", itemprop="addressLocality").text)
  except:
    city.append("null")
    
  try:
    country.append(soup.find("span", itemprop="addressCountry").text)
  except:
    country.append("null")  
  try:  
    content.append(soup.find("div", class_="content").text)
  except:
    content.append("null")
    
  founder=[]
  try:
    for name in soup.select(".name"):
      founder.append(name.text.rstrip("\n")) #.replace("\n", ",")
    founders.append(", ".join(founder))
  except:
    founders.append("null")
  
  try:
    webaddress.append(soup.select(".webaddress .content")[0].text)
  except:
    webaddress.append("null")
    
  try:
    employees.append(soup.select(".employees .content")[0].text)
  except:
    employees.append("null")    
  try:  
    looking_for.append(soup.select(".lookingfor  .content")[0].text)
  except:
    looking_for.append("null")     
    
  category = []
  try:
    for name in soup.select(".category .content a"):
      category.append(name.text.rstrip("\n")) #.replace("\n", ",")
    categorys.append(", ".join(category))
  except:
    categorys.append("null")      
    
  description = []
  try:
    for name in soup.select("#description .content"):
      description.append(name.text.rstrip("\n")) #.replace("\n", ",")
    descriptions.append(", ".join(description))
  except:
    descriptions.append("null")
  time.sleep(4)
```

```python
import pandas as pd
data = {
  "Title": title, 
  "Status":status, 
  "City": city, 
  "Country": country, 
  "Content": content, 
  "Founders": founders, 
  "WebAddress": webaddress, 
  "Employees": employees, 
  "Looking For": looking_for, 
  "Category": categorys, 
  "Descriptions": descriptions
}

df = pd.DataFrame(data)
```
