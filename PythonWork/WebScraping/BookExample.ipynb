The website to use https://toscrape.com/

Get the title of all books with a 2 star rating


```python
import requests
import bs4
```

```python
number = 1
page = f"http://books.toscrape.com/catalogue/page-{number}.html"
base_url = "http://books.toscrape.com/catalogue/page-{}.html"
```

```python
base_url.format(number)
```

```python
result = requests.get(base_url.format(number))
```

```python
soup = bs4.BeautifulSoup(result.text,"lxml")
```

```python
products = soup.select(".product_pod")
```

```python
example = products[0]
```

```python
"star-rating Three" in str(example) # the harsh way to do it
```

```python
example.select(".star-rating.Three") #select of multiple classes inside html
```

```python
# To check if there is a Two star rating class
bool(example.select(".star-rating.Two"))
```

```python
# Pick up the title of that product_pad
anchor = example.select("h3>a")[0]["title"]
```

```python
anchor
```

```python
for num in range(1,51):
    result = requests.get(base_url.format(num))
    soup = bs4.BeautifulSoup(result.text,"lxml")
    products = soup.select(".product_pod")
    
    for product in products:
        if(product.select(".star-rating.Two")):
            anchor = product.select("h3>a")[0]["title"]
            print(f"The book {anchor} has a 2 star rating \n")
        
```
