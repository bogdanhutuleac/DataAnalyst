# labelizor class how it works:

class that will take the dataframe and change categorical data to numeric

will change all columns with dtype of object into numerical, make shore you clean the data before 

first you need to define: variable_name=labelizor(df) where df is your dataframe

in order to switch the data use: variable_name.replace_categorical()

in order to switch it back use:  variable_name.retrieve_replaced()

```python
# class that will take the dataframe and change categorical data to numeric
class labelizor():
  
  labeld_dictionary = {}
  changed_columns = []
  
  def __init__(self, dataframe):
    self.dataframe = dataframe

  # create the list of changed columns
  # by adding __ in front of function name will make the function private
  def __check_object_col(self):
    columns_name = self.dataframe.columns
    object_column = []
    for name in columns_name:
      if self.dataframe[name].dtype == "object":
        object_column.append(name)
        
    # will be executed only if list is empty
    if not self.changed_columns:
      self.changed_columns = object_column
    # print(self.changed_columns)
    return self.changed_columns
  
  # take uniq values of a column and make a dictionary of values
  def __uniq_values_dict(self, series):
    uniq_values = series.unique()
    dict_values = {}
    for i,items in enumerate(uniq_values):
      dict_values[items]=i

    return dict_values
  
  # function to replace categorical
  def replace_categorical(self):
    # making a copy so it wont change the original
    new_df = self.dataframe.copy()

    list_of_columns = self.__check_object_col() # will return the list and it will store it in changed_columns as well
    replaced_dictionary = {}
    
    # if replaced dictionary doesn't exists will be created( I do this so I avoid overwrighting it by running it 2 times)
    # will be executed only if list is empty
    if not replaced_dictionary:
      for column in list_of_columns:
        replaced_dictionary[column] = self.__uniq_values_dict(new_df[column])
    
    # replace each categorical column 
    for column in list_of_columns:
      new_df[column].replace(replaced_dictionary[column], inplace=True)

    # keep a track of what changed
    self.labeld_dictionary = replaced_dictionary
    return(new_df)
    
  # retrieve categorical
  def retrieve_replaced(self):
    return self.dataframe

  
```
